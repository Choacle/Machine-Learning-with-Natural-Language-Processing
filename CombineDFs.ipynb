##Take the Dataframes

```python
import pandas as pd
df1 = pd.read_csv("Reviews.csv")

df1.columns = ['Rating','Review']
df1 = df1.sort_index()
df1
```

```python
df2 = pd.read_csv("kenaryaziBasliklari.csv")
df2
```

```python
second_list = []
import math

for i in range(len(df2["Title"])):
    if(type(df2["Title"][i]) == str):
        second_list.append(df2["Title"][i])
```

```python
df3 = pd.read_json("yemekisimleri.json", orient='columns')
df3
```

```python
yemekisimleri = []

for i in range(len(df3["Recipe"])):
    try:
        math.isnan(df3["Recipe"][i])
    except:
        yemekisimleri.append(df3["Recipe"][i]["Name"])
```

##Combine the Dataframes

```python
first_list = df1["Review"][:1000]

listofZeros = [0] * len(first_list)
listofOnes = [1] * len(second_list)
listofTwos = [2] * len(yemekisimleri)
```

```python
sumnameList = []
sumnumberList = []

sumnameList.extend(first_list)
sumnameList.extend(second_list)
sumnameList.extend(yemekisimleri)

sumnumberList.extend(listofZeros)
sumnumberList.extend(listofOnes)
sumnumberList.extend(listofTwos)
```

Clean the Texts

Cleaning the new Dataframe

```python
import re
import string

def text_temizle1(text):
    #kucuk harf yap
    text = text.lower()
    #parantez içindeki yazıları sil
    text = re.sub('\[.*?\]', '', text)
    #noktalama işaretlerini sil
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    #sayıları sil
    text = re.sub('\w*\d\w*', '', text)
    #3 karakterden kısa olan sözcükleri sil türkçe karakterlerle birlikte
    text = re.sub(r'\b[a-zğüşöçıİĞÜŞÖÇ]{1,2}\b', r'',text)
    return text
```

```python
text_clean = []

for text in sumnameList:
    text_clean.append(text_temizle1(text))
```

Dataframeyi Kaydet

```python
from pandas import DataFrame
dfClean = DataFrame({'Texts':text_clean,'Type':sumnumberList})
```

```python
dfClean.to_csv("dfNew.csv")
```
